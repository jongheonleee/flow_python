```python
class MyPoint : 
    # self 가 들어가는 이유는 해당 함수의 네임 스코프에 객체의 참조값이 없기 때문
    def __init__(self, x, y) : # 생성자
        self.x = x
        self.y = y
        
    def __str__(self) : # toString()
        return f"({self.x}, {self.y})"
    
    def __add__(self, obj) : 
        if (isinstance(obj, MyPoint) == False) :
            print("wrong data")
            return
        
        return MyPoint(self.x + obj.x, self.y + obj.y)
    
    def __repr__(self) : 
        return f"({self.x}, {self.y})"
```

```python
p = MyPoint(1, 2)
```

```python
print(p)
```

```python
p2 = MyPoint(3, 4)
```

```python
p + p2
```

```python
p + "ddd"
```
